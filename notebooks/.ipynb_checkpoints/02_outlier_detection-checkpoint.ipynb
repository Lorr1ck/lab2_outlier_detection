**Задание 2. Проверка нормальности распределения**

```python
features = ['Age', 'Fare']

for feature in features:
    data = df[feature].dropna()
    print(f"\n=== {feature} ===")
    print(f"Среднее: {data.mean():.2f}")
    print(f"Медиана: {data.median():.2f}")
    print(f"Ст. отклонение: {data.std():.2f}")
    print(f"Асимметрия: {data.skew():.2f}")
    print(f"Эксцесс: {data.kurtosis():.2f}")
    
    # QQ-plot
    fig, ax = plt.subplots()
    stats.probplot(data, dist="norm", plot=ax)
    ax.set_title(f'QQ-plot для {feature}')
    plt.show()
    
    # Тест Шапиро-Уилка
    if len(data) > 5000:
        sample = data[:5000]
    else:
        sample = data
    statistic, p_value = shapiro(sample)
    print(f"Shapiro-Wilk p-value: {p_value:.4e}")
    if p_value > 0.05:
        print("→ H0 принимается: распределение НОРМАЛЬНОЕ")
    else:
        print("→ H0 отклоняется: распределение НЕ НОРМАЛЬНОЕ")
```

**Задание 3. Обнаружение выбросов статистическими методами**


**Метод Z-score:**

```python
# Берём Age (ближе к нормальному)
age_data = df['Age'].dropna()
age_zscore = zscore(age_data)
outliers_zscore = age_data[abs(age_zscore) > 3]

print(f"=== Z-SCORE (|Z| > 3) для Age ===")
print(f"Количество выбросов: {len(outliers_zscore)}")
print(f"Доля: {len(outliers_zscore)/len(age_data)*100:.2f}%")
print(f"Значения выбросов: {sorted(outliers_zscore)}")
```

**Метод IQR:**


```python
Q1 = age_data.quantile(0.25)
Q3 = age_data.quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
outliers_iqr = age_data[(age_data < lower) | (age_data > upper)]

print(f"=== IQR для Age ===")
print(f"Q1: {Q1:.2f}, Q3: {Q3:.2f}, IQR: {IQR:.2f}")
print(f"Границы: [{lower:.2f}, {upper:.2f}]")
print(f"Количество выбросов: {len(outliers_iqr)}")
print(f"Доля: {len(outliers_iqr)/len(age_data)*100:.2f}%")
```

**Задание 4. Метод локальной плотности**

```python
# Подготовка данных
features_knn = ['Age', 'Fare', 'SibSp', 'Parch']
df_knn = df[features_knn].dropna()
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_knn)

def find_knn_outliers(data, k, threshold_percent=5):
    nn = NearestNeighbors(n_neighbors=k, metric='euclidean')
    nn.fit(data)
    distances, _ = nn.kneighbors(data)
    mean_distances = distances.mean(axis=1)
    threshold = np.percentile(mean_distances, 100 - threshold_percent)
    outliers = mean_distances > threshold
    return mean_distances, outliers, threshold

k_values = [5, 7, 9, 10]
for k in k_values:
    _, outliers, _ = find_knn_outliers(df_scaled, k)
    print(f"k={k}: аномалий = {outliers.sum()} ({outliers.sum()/len(outliers)*100:.1f}%)")
```
